```python
from tfConstrainedGauss import solve_me, random_cov_mat, InputsME, convert_mat_to_mat_non_zero
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 3

# Non-zero idxs
non_zero_idx_pairs = [(0,0),(1,1),(2,2),(1,0),(2,1)]

# Random cov mat
cov_mat = random_cov_mat(n)

plt.matshow(cov_mat)
plt.colorbar()
```

```python
target_cov_mat_non_zero = convert_mat_to_mat_non_zero(
    n=n,
    non_zero_idx_pairs=non_zero_idx_pairs,
    mat=cov_mat
)

inputs = InputsME(
    n=n,
    non_zero_idx_pairs=non_zero_idx_pairs,
    target_cov_mat_non_zero=target_cov_mat_non_zero,
    epochs=500,
    learning_rate=0.1
)
results = solve_me(inputs)
```
